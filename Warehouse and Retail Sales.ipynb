```python
import pandas as  pd;
import numpy as np;
import matplotlib.pyplot as mt;
import seaborn as sns;
```

# Remove all the null value in your data set.

```python
dd.dropna(inplace = True)
print(dd)
```

```python
dd.drop_duplicates()
print(dd)
```

```python
r =sns.histplot(data=dd,x='YEAR',hue ='ITEM TYPE',palette='prism',binwidth=0.2)
sns.move_legend(r ,"upper left",bbox_to_anchor=(1.03,1))
```

```python
m=sns.histplot(data=dd,x='MONTH',hue ='ITEM TYPE',palette='prism',binwidth=0.8)
sns.move_legend(m,"upper left",bbox_to_anchor=(1.03,1))
```

```python
rr= sns.barplot(data=dd,x='YEAR',y='RETAIL TRANSFERS',errorbar=('ci', 00),width=1.2,hue='ITEM TYPE',palette="prism_r",saturation=0.45)
sns.move_legend(rr,"upper left",bbox_to_anchor=(1.03,1))
```

```python
rt=sns.barplot(data=dd,x='MONTH',y='RETAIL TRANSFERS',errorbar=('ci', 00),width=0.6,color="m")
```

# Warehouse sales by month

```python
sales_state = dd.groupby(['WAREHOUSE SALES'], as_index=False)['MONTH'].sum().sort_values(by='MONTH', ascending=False).head(12)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'WAREHOUSE SALES',y='MONTH',color="m")
```

# Warehouse sales by year

```python
sales_state = dd.groupby(['WAREHOUSE SALES'], as_index=False)['YEAR'].sum().sort_values(by='YEAR', ascending=False).head(8)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'WAREHOUSE SALES',y='YEAR',color="m")
```

# Retail sales by year

```python
sales_state = dd.groupby(['RETAIL SALES'], as_index=False)['YEAR'].sum().sort_values(by='YEAR', ascending=False).head(8)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'RETAIL SALES',y='YEAR',color="m")
```

# Retail sales by month

```python
sales_state = dd.groupby(['RETAIL SALES'], as_index=False)['MONTH'].sum().sort_values(by='MONTH', ascending=False).head(12)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'RETAIL SALES',y='MONTH',color="m")
```
